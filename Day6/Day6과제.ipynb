최대최소의 정규화 minmax_normalize = (x-xmin)/(xmax=xmin)

```python
import numpy as np
from pandas import Series, DataFrame
import pandas as pd
from scipy.stats import zscore
```

```python
c1 = pd.read_csv('concrete.csv')
```

```python
c1.head()
```

```python
def zscore_standize(x):
    series_mean = Series(np.ones(x.size)*x.mean())
    series_std = Series(np.ones(x.size)*x.std())
    series_result = (x-series_mean)/series_std
    return series_result
```

```python
zs_concrete=c1.apply(zscore_standize)
```

```python
zs_concrete.head()
```

기존 값과 비교

```python
c1.apply(zscore).head()
```

```python
%matplotlib inline
```

```python
zs_concrete.plot()
```

```python
def minmax_normalize(x):
    return ((x-x.min())/(x.max()-x.min()))
```

```python
mm_concrete = zs_concrete.apply(minmax_normalize)
```

```python
mm_concrete.head()
```

```python
mm_concrete.plot()
```
