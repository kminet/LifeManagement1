```python
work_iter = (set(x.split(',')) for x in contacts['담당업무'])
display(work_iter)
work = sorted(set.union(*work_iter))
work
```

```python
dummies = DataFrame(np.zeros((len(contacts), len(work))), columns=work)
```

```python
dummies.tail()
```

```python
contacts.담당업무
```

```python
for i, gen in enumerate(contacts['담당업무']):
    dummies.iloc[i].loc[gen.split(',')] = 1
```

```python
dummies.head()
```

```python
contacts_work = contacts.join(dummies.add_prefix('Work_'))
contacts_work.iloc[3].head(10)
```
